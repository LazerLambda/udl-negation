```python
def fill_templates(wrd_ant_lst: list, template: str, mask_token: str, editor: checklist.editor.Editor) -> list[tuple[str, str]]:
    return_list: list = []
    for (wrd, ant) in wrd_ant_lst:
        # x0: str = template.format(word=wrd, antonym=ant)
        # y0: str = template.format(word=wrd, antonym=mask_token)
        # x1: str = template.format(word=wrd, antonym=ant)
        # y1: str = template.format(word=mask_token, antonym=ant)
        main, maksed0, masked1 = editor.template(template, word=[wrd, '<mask>'], antonym=[ant, '<mask>']).data[0:3]
        return_list.append((main, maksed0))
        return_list.append((main, masked1))
    return return_list
```
