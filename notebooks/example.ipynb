```python
import transformers
from transformers import pipeline, AutoModelForMaskedLM, AutoTokenizer
from transformers.modeling_utils import load_state_dict
import torch
```

```python
unmasker = pipeline('fill-mask', model="prajjwal1/bert-small")

unmasker("A robin is not a [MASK].")
```

```python
model_str = "/home/philko/Documents/Uni/WiSe2223/UnsupervisedLearning/udl-negation/models/final_model_large.pt"
```

```python
load_state_dict(model_str)
```

```python
model: AutoModelForMaskedLM = AutoModelForMaskedLM.from_pretrained("prajjwal1/bert-small", state_dict=load_state_dict(model_str))
```

```python
tokenizer: AutoTokenizer = AutoTokenizer.from_pretrained(
            "prajjwal1/bert-small",
            use_fast=True,
            max_len=128)
```

```python
tokenizer.save_pretrained(model_str)
```

```python
unmasker = pipeline('fill-mask', model=model, tokenizer=tokenizer)

unmasker("A robin is not a [MASK].")
```
