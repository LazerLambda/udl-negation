```python
path: str = "/home/philko/Documents/Uni/WiSe2223/UnsupervisedLearning/udl-negation/data/processed/roberta_filtered/negation_aware_data.txt"
f = open(path, 'r')
counter: int = 0
for line in f:
    counter += 1
f.close()
indices = np.random.choice(range(counter), 755200, replace=False)  # 83, replace=False)
indices.sort()
indices = list(indices)
f = open(path, 'r')
collector: list = []
counter = 0
for i, text in enumerate(f):
    if i == indices[counter]:
        collector.append(text)
        if counter >= len(indices) - 1:
            break
        else:
            counter += 1
f.close()
```

```python
df_orig: pd.DataFrame = pd.DataFrame({'text': collector})
df_orig
```

```python
df_synth: pd.DataFrame = pd.read_csv('/home/philko/Documents/Uni/WiSe2223/UnsupervisedLearning/udl-negation/data/processed/wn_neg_processed/data_wn_masked.txt', header=None)
df_synth.columns = ['text']
df_synth.text = list(map(lambda e: str(e) + '\n', df_synth.text.values))
df_synth
```

```python
df: pd.DataFrame = pd.concat([df_orig, df_synth])
df = df.sample(frac=1).reset_index(drop=True)
```
